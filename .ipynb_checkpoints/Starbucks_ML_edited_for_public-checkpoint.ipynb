```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline  

import scipy.stats as stats
```

```python
#load data from csv file
data = pd.read_csv("WHERE DATA FILE GOES")


# create a failures dataframe that stores average failures, standard deviation of values, failure proportion average,
# faiure proportion standard deviation, and count ALL BY SOURCE
failures = data.groupby(data.src).mean()
prop = failures.proportion.values
failures = failures[['failure']]

size = data.groupby(data.src).size()
stdev = data.groupby(data.src).std()

failures['failure_stdev'] = stdev.failure.values
failures['prop_avg'] = prop
failures['prop_stdev'] = stdev.proportion.values
failures['count'] = size.values

print data.describe()
print
print
print failures.describe()
```

```python
dict2= {
    'A change has been made to Windows Firewall exception list. A rule was added': [],
    'A change has been made to Windows Firewall exception list. A rule was deleted': [],
    'A computer account was changed': [],
    'A member was added to a security-enabled global group': [],
    'A member was added to a security-enabled universal group': [],
    'A network share object was accessed': [],
    'A privileged service was called': [],
    'A user account was created': [],
    "An attempt was made to change an account's password": [],
    'An attempt was made to reset an accounts password': [],
    'An operation was attempted on a privileged object': [],
    'Special privileges assigned to new logon': [],
    'User name does not exist': [],
    'User name is correct but the password is wrong': []
}

for item in data.signature.values:
    for key in dict2.keys():
        if key in item:
            dict2[key].append(1)
        else:
            dict2[key].append(0)

for key in dict2.keys():
    data[key] = dict2[key]
```

```python
dict3={}

for item in data.brute_force_concept.values:
    if item not in dict3.keys():
        dict3[item] = []
        
for item in data.brute_force_concept.values:
    for key in dict3.keys():
        if key==item:
            dict3[key].append(1)
        else:
            dict3[key].append(0)

for key in dict3.keys():
    data[key] = dict3[key]
    
data
```

```python
#calculate the probability of observed failed logins given Poisson distribution and z_score for failure proportions

prob_z_score = []
failure_poisson_prob = []

for i in range(len(data)):
    if data.src.values[i] in failures.index:
        lamb = failures[failures.index==data.src.values[i]].failure.values[0]
        k = data.failure.values[i]
        poisson = stats.distributions.poisson.pmf(k, lamb)
        failure_poisson_prob.append(poisson)

        x = data.proportion.values[i]
        mean = failures[failures.index==data.src.values[i]].prop_avg.values[0]
        std = failures[failures.index==data.src.values[i]].prop_stdev.values[0]
        if np.isnan(std) or std==0.0:
            std = 0.1
        prob_z_score.append((np.absolute((x-mean)))/std)
    else:
        print 'found one'
        lamb = failures.failure.values.mean()
        k = data.failure.values[i]
        poisson = stats.distribution.poisson.pmf(k,lamb)
        failure_poisson_prob.append(poisson)
        
        x = data.proportion.values[i]
        mean = data.proportion.values.mean()
        std = data.proportion.values.std()
        if np.isnan(std) or std==0.0:
            std = 0.1
        prob_z_score.append((np.absolute((x-mean)))/std)

data['z_score'] = prob_z_score
data['failure_poisson_prob'] = failure_poisson_prob
```

```python
fields = ['src', 'auth_attempts', 'count_auth_apps', 'dest_count', 'failure', 'hour', 'proportion', 'success', 
          'unique_users_count', 'z_score', 'failure_poisson_prob'] + dict2.keys() + dict3.keys()
model_data = data[fields]
print model_data.columns

model2 = model_data[['src','failure', 'proportion', 'z_score', 'failure_poisson_prob', 'high', 'medium', 'extreme',
                    'auth_attempts', 'dest_count']]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn.decomposition import PCA


train, test = train_test_split(model_data, test_size=0.33, random_state=0)

src = train.src
train_standardized = train.drop(['src'], axis=1)

#standardize the features data
train_standardized = pd.DataFrame(StandardScaler().fit_transform(train_standardized))


components = range(1,20)
explained_variance = []
for i in components:
    pca = PCA(n_components=i)
    pca = pca.fit(train_standardized)
    explained_variance.append(sum(pca.explained_variance_ratio_))
    
#change the figure size of the plot
plt.figure(figsize=(8,6))
#change the axis range for both x and y axis
plt.axis([0,21,.2,1.005])
#use subplot to add functionality
ax = plt.subplot(111)

#plot components to level of variance explained at each component level 
plt.plot(components, explained_variance, color="#00704a", lw=2, marker='h')
plt.xlabel('Number of Components', weight='bold')
plt.ylabel('Amount of Variance Explained by Components', weight='bold')
plt.title('Explained Variance by Number of Components\n', weight='bold', size=15)
plt.grid(True)

#remove top and right axis and ticks 
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
plt.show()
print explained_variance
```

```python
from sklearn.cluster import DBSCAN

pca = PCA(n_components=15)
pca = pca.fit_transform(train_standardized)

pca_df =pd.DataFrame(pca)

dbscan = DBSCAN(random_state=111, min_samples=100)
dbscan.fit(pca_df)
np.unique(dbscan.labels_)
```
